```python
import matplotlib as plt

print("Matplotlib version:", plt.__version__)
```

```python
# Line Plots
x = [1, 2, 3, 4, 5]
y = [10,20, 25,30, 40]
plt.pyplot.plot(x, y, label='Line 1', color='b', marker='o')
plt.pyplot.title('Line Plot Example')
plt.pyplot.xlabel('X-axis')
plt.pyplot.ylabel('Y-axis')
plt.pyplot.legend()
plt.pyplot.grid()
plt.pyplot.show()
```

```python
# Bar Plots
names = ["Amit", "Riya","Ram","Sana"]
marks = [78,85,80,90]
plt.pyplot.bar(names, marks, color='cyan', label='Marks')
plt.pyplot.title('Bar Plot Example')
plt.pyplot.legend()
plt.pyplot.xlabel('Names')
plt.pyplot.ylabel('Marks')
plt.pyplot.show()
```

```python
# Histograms Plots
ages = [21,22,23,21,25,26,22,24,23,22,26,25]
plt.pyplot.hist(ages, edgecolor='black', bins=10, color='green', alpha=0.7)
plt.pyplot.title('Histogram Example')
plt.pyplot.xlabel('Age Groups')
plt.pyplot.ylabel('Number of People')
plt.pyplot.show()
```

```python
# With Pandas
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("student.csv")
plt.bar(df['Name'], df['Marks'], color='orange')
plt.title('Student Marks')
plt.xlabel('Names')
plt.ylabel('Marks')
plt.show()
```
