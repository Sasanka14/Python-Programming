```python
import pandas as pd
import matplotlib as plt
```

```python
df = pd.DataFrame({
    "Year" : [2018,2019,2020,2021,2022],
    "Sales" : [200,250,230,300,320]
})
```

```python
# Line Plot
plt.pyplot.plot(df['Year'], df['Sales'], marker='o', linestyle='-', color='b')
plt.pyplot.title('Year Over Sales')
plt.pyplot.xlabel('Year')
plt.pyplot.ylabel('Sales')
plt.pyplot.grid(True)
plt.pyplot.show()
```

```python
# Bar Plot
plt.pyplot.bar(df['Year'], df['Sales'], color='green')
plt.pyplot.title('Year-wise Sales Comparison')
plt.pyplot.xlabel('Year')
plt.pyplot.ylabel('Sales')
plt.pyplot.show()
```

```python
# Histogram
plt.pyplot.hist(df['Sales'], bins=5, color='orange', edgecolor='black')
plt.pyplot.title('Sales Distribution')
plt.pyplot.xlabel('Sales')
plt.pyplot.ylabel('Frequency')
plt.pyplot.show()
```

```python
# Scatter Plot
plt.pyplot.scatter(df['Year'], df['Sales'], color='red')
plt.pyplot.title('Year vs Sales Scatter Plot')
plt.pyplot.xlabel('Year')
plt.pyplot.ylabel('Sales')
plt.pyplot.show()
```
